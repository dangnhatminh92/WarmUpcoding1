```python
import pandas as pd
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python

# Bc 1: c d liu
data = """
T xa n nay, dn tc Vit Nam lun coi trng ngn ng ca mnh - ting Vit. Chng ta c 
th t ho m khng nh rng ting Vit l mt th ting p. Ting Vit nc ta rt giu v 
p. N l th ting hi ha v m hng, thanh iu. Bi vy, ting Vit c y  kh nng 
truyn ti t tng, tnh cm ca ngi Vit Nam nhm p ng nhu cu trong i sng ngi 
Vit. N l th ting giu cht nhc bi n l th ting giu thanh iu. T vng ting Vit qua 
cc thi k din bin tng ln nhiu, ng php uyn chuyn v chnh xc hn. Bi vy, con 
ngi Vit Nam hm nay cn bit gn gi v pht huy s trong sng, giu p ca ting Vit.
"""

# Bc 2: Tin x l vn bn
with open('vietnamese-stopwords.txt', 'r', encoding='utf-8') as f:
    stopwords = set(line.strip() for line in f if line.strip())
def clean_text(text):
    text = text.lower() # chuyn v ch thng
    text = re.sub(r'\s+', ' ', text).strip() # loi b khong trng tha
    text = re.sub(r'\d+', '', text) # loi b cc s
    text = text.replace('\r', ' ').replace('\n', ' ') # loi b cc k t xung dng
    text = re.sub(r'[^a-zA-Z\s]', '', text) #Gi li cc k t ch ci v khong trng
    text = ' '.join([word for word in text.split() if word not in stopwords]) # loi b cc t stopwords
    return text
 
text = clean_text(data)

# Bc 3: To v hin th WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud t d liu  x l', fontsize=16)
plt.show()
```
