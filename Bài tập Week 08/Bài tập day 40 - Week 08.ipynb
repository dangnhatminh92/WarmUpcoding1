```python
def compute_cosine_similarity(vector1, vector2):
    """
    Tính toán độ tương tự cosine giữa hai vector.
    
    Args:

    """
    vector1 = vector1.values
    vector2 = vector2.values
    numerator = np.dot(vector1, vector2)
    denominator = np.linalg.norm(vector1) * np.linalg.norm(vector2)
    cosine_similarity_value = numerator / denominator
    return cosine_similarity_value


```

```python
vector1 = df['Total Fat']
vector2 = df['Calories']
print(vector1)
compute_cosine_similarity(vector1, vector2)
cosine_similarity_value = compute_cosine_similarity(vector1, vector2)
print(f"Độ tương tự cosine giữa hai vector là: {cosine_similarity_value}")
```

```python
#Biểu diên Sodium
sns.boxplot(data=df, y='Sodium')
plt.title('Biểu đồ hộp Sodium')
plt.xlabel('Sodium (mg)')
plt.tight_layout()
plt.show()


```

```python
data = df['Sodium']

def find_outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return lower_bound, upper_bound

lower_bound, upper_bound = find_outliers(data)


data_outliers = df[(df['Sodium'] > upper_bound) | (df['Sodium'] < lower_bound)]
print(f"Giá trị outlier:\n{data_outliers['Sodium']}")
```
